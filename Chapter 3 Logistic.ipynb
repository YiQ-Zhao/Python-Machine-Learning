# Modeling class probabilities via logistic regression

## Logistic regression intuition and conditional probabilities  
$$logit(p) = log\frac{p}{(1-p)}$$  
The logit function takes input values in the range 0 to 1 and transforms them to values over the entire real number range, which we can use to express a linear relationship between feature values and the log-odds:
$$logit(p(y=1\mid \mathbf{x}))=w_0x_0 + w_1x_1+...+w_mx_m=\sum^m_{i=0}w_ix_i=\mathbf{w}^T\mathbf{x}$$  
Logistic function, sometimes simply abbreviated as sigmoid function due to its characteristic S-shape  
$$\phi(z)=\frac{1}{1+e^{-z}}$$

In Adaline, we used the identify function $\phi(z)=z$ as the activation function. In logistic regression, this activation function simply becomes the sigmoid function that we defined earlier, which is illustrated in the following figure:

## Learning the weights of the logistic cost function
To explain how we can derive the cost function for logistic regression, we first need to define the likelihood $L$ that we want to maximize when we build a logistic regression model, assuming that the individual samples in our dataset are independent of one another. The formula is as follows:
$$L(\mathbf{w}) = P(\mathbf{y\mid x;w})=\prod^{n}_{i=1}P(y^{i}\mid x^{(i)};\mathbf{w})=\prod_{i=1}^n=\prod_{i=1}^{n}(\phi(z^{(i)}))^{y^{(i)}}(1-\phi(z^{(i)}))^{1-y^{(i)}}$$  
In practice, it is easier to maximize the (natural) log of this equation, which is called the log-likelihood function:  
$$l(\mathbf{w})=logL(\mathbf{w})=\sum_{i=1}^n[y^{(i)}log(\phi(z^{(i)}))+(1-y^{(i)})log(1-\phi(z^{(i)}))]$$

Now we could use an optimization algorithm such as gradient ascent to maximize this log-likelihood function. Alternatively, let's rewrite the log-likelihood as a cost function $J$ that can be minimized using gradient descent in *Chapter 2, Training Machine Learning Algorithm for Classification:*
$$J(\mathbf{w})=\sum^{n}_{i=1}[-y^{(i)}log(\phi(z^{(i)}))-(1-y^{(i)})log(1-\phi(z^{(i)}))]$$
